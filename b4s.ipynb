```python
import pandas as pd
```

```python
def search_job(time):
    for i in range(1, 51):
        for job in list_job.find_all("div", {"data-job-position": f"{i}"}):
            position = job.find("span").string
            company = job.find("a", class_="company").string
            location = job.find("label", class_="address truncate").string.strip()
            updated_time = job.find("label", class_="address mobile-hidden label-update").get_text().strip()
            if(updated_time.find(time)!=-1):
                print(f"Position: {position}")
                print(f"Company: {company}")
                print(f"Location: {location}")
                print(f"Updated Time: {updated_time}")
                print("-" * 40) 
```

```python
if __name__ == '__main__':
    time = input("Please give me the time (giờ/ngày/tuần) company publish the job: ")
    print(search_job(time))
```
